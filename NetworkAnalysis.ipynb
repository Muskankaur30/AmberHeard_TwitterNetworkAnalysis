```python
df = pd.read_csv("2017-2022 Twitter all_quotes retweets Amber Heard Timeline_Data.csv",
                 usecols = ["referenced_tweets", "id", "text", "public_metrics.retweet_count","public_metrics.reply_count","quoted_tweet.public_metrics.like_count","quoted_tweet.public_metrics.quote_count","quoted_tweet.id", "quoted_tweet.author.name", "author.id","author.name","author.public_metrics.followers_count","author.public_metrics.followers_count","author.username","author.created_at","entities.hashtags", "author.url"])
```

```python
df.head()
df.tail()
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['quoted_tweet.public_metrics.like_count'].fillna(df['quoted_tweet.public_metrics.like_count'].mean(), inplace=True)
df['quoted_tweet.public_metrics.quote_count'].fillna(df['quoted_tweet.public_metrics.quote_count'].mean(), inplace=True)
```

```python
df['author.url'].fillna("Not Known", inplace=True)
df['entities.hashtags'].fillna("Not Known", inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['author.created_at']
```

```python
df['date'] = pd. to_datetime(df['author.created_at'])
```

```python
df['date']
```

```python
df['year'] = pd.DatetimeIndex(df['date']).year
```

```python
dfsample = df.loc[df['year'] == 2020, :]
```

```python
u_user = dfsample['author.id'].unique()
```

```python
u_user.sort()
u_code = ["u" + str(i) for i in range(len(u_user)) ]
u_code[:5]
```

```python
map_dict = dict(zip(u_user, u_code))
```

```python

list(map_dict.items())[:5]
```

```python
dfsample["author.id"] = dfsample["author.id"].replace(map_dict, inplace= False)
```

```python
dfsample.head()
```

```python
dfsample['quoted_tweet.id'].value_counts()
```

```python
quotedId = dfsample['quoted_tweet.id'].unique()
```

```python
quotedId.sort()
```

```python
quoted_c = ["qID" + str(i) for i in range(len(quotedId)) ]
map_dic2 = dict(zip(quotedId, quoted_c))
```

```python
dfsample['quoted_tweet.id'] = dfsample['quoted_tweet.id'].replace(map_dic2, inplace= False)
```

```python
dfsample.shape
```

```python
dfsample.head()
```

User Nodes

```python
grpd_user = dfsample.groupby(['author.id'])
user_nodes = grpd_user.agg({'public_metrics.retweet_count' : [('u_min','min'),('u_max','max'),('u_mean','mean'),('u_std',np.std)] }).reset_index()
user_nodes = user_nodes.round(decimals = 2)
user_nodes.head()
```

```python
dfsample['is_retweeted'] = np.where(dfsample['public_metrics.retweet_count'] != 0, True, False)
```

```python
user_nodes['Cat'] = 'Twitter_User'
```

```python
user_nodes['is_retweet'] = dfsample['is_retweeted']
users_retweet = dfsample.loc[dfsample["is_retweeted"] == 1, 'author.id' ].unique()
len(users_retweet)
```

```python
user_nodes['retweet'] = dfsample.loc[dfsample["is_retweeted"] , 'author.id' ]
```

```python
user_nodes['Label'] = user_nodes['author.id']
```

```python
user_nodes = user_nodes.rename(columns = {'User': "Id"})
```

```python
user_nodes.head()
```

```python
user_nodes['is_retweet'] = user_nodes['is_retweet'].map({"True":1 , "False" : 0})
```

```python
user_nodes.head()
```

```python
user_nodes.to_csv("Twitter_users.csv",
                  index = False,
                  sep = ";"
                  )
```

Quoted ID

```python
grpd_quoted = dfsample.groupby(['quoted_tweet.id'])
grpd_quoted = grpd_quoted.agg({'public_metrics.retweet_count' : [('q_min','min'),('q_max','max'),('q_mean','mean'),('q_std',np.std)] }).reset_index()
grpd_quoted = grpd_quoted.round(2)
grpd_quoted.head()
grpd_quoted.shape
```

```python
grpd_quoted['Cat'] = 'quoted_tweet'
```

```python
grpd_quoted['retweet'] = dfsample.loc[dfsample["is_retweeted"] , 'quoted_tweet.id' ]
```

```python
grpd_quoted['Label'] = grpd_quoted['quoted_tweet.id']
grpd_quoted = grpd_quoted.rename(columns = {'Merchant Name': "Id"})
grpd_quoted.head()
grpd_quoted.shape
```

```python
grpd_quoted['is_retweeted'] = grpd_quoted['is_retweeted'].map({True:1 , False : 0})
```

```python
grpd_quoted.to_csv("Quoted_Tweets.csv",
                      index = False,
                      sep = ";")
```

Edges

```python
edges = dfsample.groupby([dfsample['author.id'], dfsample['quoted_tweet.id']]).size()
```

```python
edges = dfsample[['author.id', 'quoted_tweet.id']].groupby(['author.id', 'quoted_tweet.id']).size().reset_index()
edges = edges.rename(columns = { 0 : "weight"})
edges.head()
```

```python
edges = edges.rename(columns = {'author.id': 'Source', 'quoted_tweet.id': 'Target'})
```

```python
edges['Type']= 'Undirected'
edges.head()
```

```python
edges.to_csv("twitter_edges.csv",
             index = False,
             sep = ";"
             )
```
